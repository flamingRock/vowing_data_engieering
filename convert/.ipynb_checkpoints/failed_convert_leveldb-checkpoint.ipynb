```python
import pandas as pd
import subprocess
import json
import os
import plyvel
import sys
from google.protobuf.json_format import MessageToJson
from google.protobuf.json_format import MessageToDict
sys.path.append("C:\\Users\\admin")
import document_pb2
```

```python
# leveldb_file_path = "D:/DATA_PREPROCESS/firestore_datas/2023-08-10T09_07_44_34692/all_namespaces/kind_Attend/"
leveldb_file_path = "D:/DATA_PREPROCESS/FIRESTORE_DATAS/2023-10-19T02_41_53_60252/all_namespaces/kind_Attend/"

json_output_folder = os.path.join(leveldb_file_path, "JSON")
if not os.path.exists(json_output_folder):
    os.makedirs(json_output_folder)

for i in range(192):
    file_path = os.path.join(leveldb_file_path, f'output-{i}')
    firestore_data = document_pb2.Document()

    try:
        # 파일에서 데이터 읽기
        with open(file_path, 'rb') as f:
            # data_bytes = f.read()
            firestore_data.ParseFromString(f.read())
        
        # 변환한 데이터 확인
        converted_data = MessageToDict(firestore_data, preserving_proto_field_name=True)
        print(f"Converted data for file output-{i}:", converted_data)
        
        # 디코딩된 데이터를 Python 딕셔너리로 변환
        data_dict = json.loads(json.dumps(converted_data))

        # JSON 파일로 저장
        json_output_path = os.path.join(json_output_folder, f'output-{i}.json')
        with open(json_output_path, 'w') as json_file:
            json.dump(data_dict, json_file, indent=2)
        
    except Exception as e:
        print(f"Error processing file output-{i}: {e}")
#         # 일부 데이터를 출력하여 어떤 내용이 문제를 일으키는지 확인
#         print(data_bytes[:100])  # 처음 100바이트만 출력
```

```python
# protoc이 설치되어 있는지 확인
try:
    output = subprocess.check_output(['protoc', '--version'])
    print(f"Protoc version: {output.decode('utf-8')}")
except FileNotFoundError:
    raise FileNotFoundError("protoc이 발견되지 않았습니다. 설치되어 있고 PATH에 있어야 합니다.")
```

```python
# 파일 경로 설정
base_dir = "D:/DATA_PREPROCESS/FIRESTORE_DATAS/Users_Attend/Users_Attend_231004/all_namespaces/all_kinds/"
output_file = 'decoded_metadata'

# 디렉토리 내의 모든 파일을 수집
files = [f for f in os.listdir(base_dir) if os.path.isfile(os.path.join(base_dir, f))]

decoded_data_all = []

for file in files:
    input_path = os.path.join(base_dir, file)
    
    # 파일을 디코딩하기 위해 protoc 사용
    with open(input_path, 'rb') as f:
        try:
            decoded_data = subprocess.check_output(['protoc', '--decode_raw'], input=f.read())
            decoded_data_all.append(decoded_data.decode('utf-8'))
        except subprocess.CalledProcessError as e:
            print(f"{file} 파일을 디코딩하는 중 오류 발생: {e}")

# 여기서 decoded_data_all을 파싱하여 원하는 JSON 형식으로 변환할 수 있습니다.
# 이 예제에서는 그냥 모두 합친 다음 저장합니다.

output_path = os.path.join(base_dir, f"{output_file}.json")

with open(output_path, 'w', encoding='utf-8') as json_file:
    json_file.write("\n".join(decoded_data_all))
```
