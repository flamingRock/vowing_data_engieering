```python
# 파일 경로 설정
base_dir = "D:/DATA_PREPROCESS/FIRESTORE_DATAS/2023-08-10T09_07_44_34692/all_namespaces/kind_Attend"
output_file = 'decoded_metadata'

# protoc이 설치되어 있는지 확인
try:
    subprocess.check_output(['protoc', '--version'])
except FileNotFoundError:
    raise FileNotFoundError("protoc이 발견되지 않았습니다. 설치되어 있고 PATH에 있어야 합니다.")

# 디렉토리 내의 모든 파일을 수집
files = [f for f in os.listdir(base_dir) if os.path.isfile(os.path.join(base_dir, f))]

decoded_data_all = []

for file in files:
    input_path = os.path.join(base_dir, file)
    
    # 파일을 디코딩하기 위해 protoc 사용
    with open(input_path, 'rb') as f:
        try:
            decoded_data = subprocess.check_output(['protoc', '--decode_raw'], input=f.read())
            decoded_data_all.append(decoded_data.decode('utf-8'))
        except subprocess.CalledProcessError as e:
            print(f"{file} 파일을 디코딩하는 중 오류 발생: {e}")

# 여기서 decoded_data_all을 파싱하여 원하는 JSON 형식으로 변환할 수 있습니다.
# 이 예제에서는 그냥 모두 합친 다음 저장합니다.

output_path = os.path.join(base_dir, f"{output_file}.json")

with open(output_path, 'w', encoding='utf-8') as json_file:
    json_file.write("\n".join(decoded_data_all))
```
