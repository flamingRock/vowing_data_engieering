```python
point_db = db.collection("Point")

# 기준 날짜 설정 ('23.11.14')
base_date_str = '23.11.14'
base_date = datetime.datetime.strptime(base_date_str, '%y.%m.%d')

# 쿼리 수행: 'date' 필드의 시작 날짜가 기준 날짜 이후인 문서 10개 검색
query = point_db.where('date', '>=', base_date_str).limit(10)
documents = query.stream()

# 각 문서의 필드 데이터 타입 출력
for doc in documents:
    print(f'문서 ID: {doc.id}')
    for field, value in doc.to_dict().items():
        print(f'  필드: {field}, 데이터 타입: {type(value).__name__}')
    print('---')
```
