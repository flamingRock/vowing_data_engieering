```python
import os
```

```python
def list_files(startpath):
    graphic_structure = ""
    for root, dirs, files in os.walk(startpath):
        level = root.replace(startpath, '').count(os.sep)
        indent = ' ' * 4 * (level)
        graphic_structure += f"{indent}{os.path.basename(root)}/\n"
        subindent = ' ' * 4 * (level + 1)
        for f in files:
            graphic_structure += f"{subindent}{f}\n"
    return graphic_structure
```

```python
# 시작 경로 설정
startpath = 'D:/DATA_PREPROCESS/FIRESTORE_DATAS/sample_voice_metadata_230918/ETRI_VOICE_DATASET'
```

```python
# 폴더와 파일 구조를 그래픽하게 출력
graphic_structure = list_files(startpath)
print(graphic_structure)
```
