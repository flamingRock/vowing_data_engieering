```python
count_with_values_of_local = df['local_code'].count()
count_missing_values_of_local = len(df) - count_with_values_of_local
print(f"local_code에 값이 있는 행의 개수: {count_with_values_of_local}")
print(f"local_code에 값이 없는 행의 개수: {count_missing_values_of_local}")
```

```python
count_test = df['is_test'].value_counts().get(1, 0)
count_not_test = df['is_test'].value_counts().get(0, 0)
print(f"is_test에 값이 0인 행의 개수: {count_not_test}")
print(f"is_test에 값이 1인 행의 개수: {count_test}")
```

```python
gender_counts = df['gender'].value_counts()

labels = ['M', 'F']
sizes = [gender_counts['M'], gender_counts['F']]

# 원형 차트
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # 원형 차트가 원 모양이 되도록 합니다.
plt.title('Gender Distribution')
plt.show()
```

```python
# 'birth_year' 열의 값을 범위별로 분류합니다.
birth_year_bins = [0, 59, 69, 79, 89, 99, 2009, float('inf')]
birth_year_labels = ['under 59', 'under 69', 'under 79', 'under 89', 'under 99', 'under 09', 'over 10']
birth_year_categories = pd.cut(df['birth_year'], bins=birth_year_bins, labels=birth_year_labels)

# 각 범위의 개수를 계산합니다.
birth_year_counts = birth_year_categories.value_counts()

# 원형 차트에 사용할 라벨과 값입니다.
labels = birth_year_labels
sizes = [birth_year_counts[label] for label in birth_year_labels]

# 막대그래프 그리기
plt.bar(labels, sizes)
plt.title('Age Group Distribution')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()
```

```python
# birth_year에 따라 카테고리 분류
df['age_group'] = 'Elderly'
df.loc[df['birth_year'] >= 59, 'age_group'] = 'Senior'
df.loc[df['birth_year'] >= 74, 'age_group'] = 'Middle-aged'
df.loc[df['birth_year'] >= 89, 'age_group'] = 'Young'

# 각 카테고리의 개수 계산
age_group_counts = df['age_group'].value_counts()

# 원형 차트 그리기
labels = age_group_counts.index
sizes = age_group_counts.values
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # 원형 차트가 동그랗게 그려지도록 함
plt.title('Age Group Distribution')
plt.show()
```

```python
# local_code 값의 빈도를 계산 (NaN 값도 포함)
local_code_counts = df['local_code'].fillna('Non').value_counts()

# 지역 코드와 해당 레이블을 매핑 (Non 추가)
local_code_labels = {
    'Je': 'Jeue',
    'Se': 'Seoul',
    'Gy': 'Gyeonggi',
    'Ch': 'Chungcheong',
    'Jd': 'Jeolla',
    'Ga': 'Gangwon',
    'Non': 'Non'
}

# 레이블 순서에 따라 값을 정렬
local_code_values = [local_code_counts.get(code, 0) for code in local_code_labels.keys()]
local_code_names = list(local_code_labels.values())

# 전체 카운트 대비 비율 계산
total_count = sum(local_code_values)
local_code_percentages = [(value / total_count) * 100 for value in local_code_values]

local_code_names = list(local_code_labels.values())

# 막대 그래프 생성
colors = ['blue', 'green', 'red', 'purple', 'orange', 'brown', 'gray']
plt.bar(local_code_names, local_code_percentages, color=colors)

# 축 레이블과 제목 추가
plt.xlabel('Local Code')
plt.ylabel('Percentage (%)')
plt.title('Distribution of Local Codes')

# 그래프 표시
plt.show()
```

```python
# record_time을 문자열로 변환 (이미 문자열인 경우 필요 없음)
df['record_time'] = df['record_time'].astype(str)

# record_time이 NaN인 행을 제외한 새 데이터프레임 생성
record_time_df = df.dropna(subset=['record_time'])

# 연, 월, 일, 시, 분, 초를 추출
record_time_df['year'] = record_time_df['record_time'].str[:4]
record_time_df['month'] = record_time_df['record_time'].str[4:6]
record_time_df['day'] = record_time_df['record_time'].str[6:8]
record_time_df['hour'] = record_time_df['record_time'].str[8:10]
record_time_df['minute'] = record_time_df['record_time'].str[10:12]
record_time_df['second'] = record_time_df['record_time'].str[12:]

# 연과 월을 기준으로 개수 계산
monthly_counts = record_time_df.groupby(['year', 'month']).size().reset_index(name='counts')

# 달별로 플롯
for year in monthly_counts['year'].unique():
    subset = monthly_counts[monthly_counts['year'] == year]
    plt.plot(subset['month'], subset['counts'], label=f'Year {year}', marker='o')

# 축 레이블과 제목 추가
plt.xlabel('Month')
plt.ylabel('Counts')
plt.title('Monthly Added Voice Data Distribution')

# 범례 추가
plt.legend()

# 그래프 표시
plt.show()
```
