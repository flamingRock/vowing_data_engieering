```python
import pandas as pd
import json
import os
import plyvel
import sys
sys.path.append("C:\\Users\\admin")
import document_pb2

# print(sys.path)
```

```python
json_file_path = "D:/DATA_PREPROCESS/firestore_datas/2023-08-10T09_07_44_34692/all_namespaces/kind_Attend/"
# db_path = os.path.join(json_file_path, 'output-1')  # LevelDB 디렉토리 경로 설정

json_output_folder = os.path.join(json_file_path, "JSON")
if not os.path.exists(json_output_folder):
    os.makedirs(json_output_folder)

for i in range(192):  # 0부터 191까지
    db_path = os.path.join(json_file_path, f'output-{i}')
    
    # LevelDB를 연다
    db = plyvel.DB(db_path, create_if_missing=False)
    
    for key, value in db:
        # 데이터를 디코딩하기 위한 객체 초기화
        firestore_data = document_pb2.Document()
        firestore_data.ParseFromString(value)

        # 디코딩된 데이터를 Python 딕셔너리로 변환
        data_dict = json.loads(json.dumps(firestore_data, indent=2, default=str))

        # JSON 파일로 저장
        json_output_path = os.path.join(json_output_folder, f'{key.decode("utf-8")}.json')
        with open(json_output_path, 'w') as json_file:
            json.dump(data_dict, json_file, indent=2)

    db.close()

#------------------------------------------------------------------------------------------------------------------------------------#


# # 적절한 청크 크기 찾기
# for chunk_size in [100, 500, 1000, 5000, 10000, 50000, 100000]:
#     try:
#         chunk = next(pd.read_json(file_path, lines=True, chunksize=chunk_size))
#         print(f"청크 크기 {chunk_size}로 읽기 성공")
#         continue
#     except MemoryError:
#         print(f"청크 크기 {chunk_size}로 읽기 실패 - 메모리 오류")

# chunk_size = 50000 # 원하는 청크 사이즈, 50000행
# chunks = []
# total_rows_read = 0
# for chunk in pd.read_json(file_path, lines=True, chunksize=chunk_size):
#     chunks.append(chunk)
#     total_rows_read += len(chunk)
#     print(f"현재까지 읽은 행의 수: {total_rows_read}")

# # 첫 청크의 처음 5개 행 출력
# print(chunks[0].head())
```

```python
# 여러 인코딩을 시도하여 파일을 읽어보고, 어느 인코딩으로 읽혔는지와 파일의 데이터 형식을 확인
file_data = {}

def check_file_type(file_content):
    try:
        json.loads(file_content)
        return "JSON"
    except:
        pass

    # 다른 파일 형식에 대한 확인을 여기에 추가할 수 있습니다.
    return "UNKNOWN"

# 인코딩을 시도하고 파일 타입을 체크하는 함수
def try_encodings(file_path):
    encodings_to_try = ["utf-8", "utf-16", "cp949", "latin2"]
    for enc in encodings_to_try:
        try:
            with open(file_path, "r", encoding=enc) as file:
                content = file.readlines()[:10]  # 처음 10줄만 가져옴
                content_str = ''.join(content)
                
                file_type = check_file_type(content_str)
                
                if file_type:
                    return enc, file_type, content_str
        except:
            continue
    return None, "UNKNOWN", None

# 0부터 191까지의 output 파일을 읽음
for i in range(192):  
    file_path = os.path.join(json_file_path, f"output-{i}")

    if os.path.exists(file_path):
        encoding, file_type, content = try_encodings(file_path)
        file_data[file_path] = {
            'encoding': encoding,
            'type': file_type,
            'sample_content': content
        }

for path, data in file_data.items():
    print(f"File: {path}")
    print(f"Encoding: {data['encoding']}")
    print(f"File Type: {data['type']}")
    print(f"Sample Content: {data['sample_content']}\n{'-'*50}\n")
```
